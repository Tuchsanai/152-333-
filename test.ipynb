```
import numpy as np
from sklearn.neighbors import NearestNeighbors
```

```
X = np.random.randint(0,10,(100,2))
```

```
X[0:10]
```

```
model =  NearestNeighbors(n_neighbors=3)
model.fit(X)
```

```
distance,index = model.kneighbors(X)
```

```
idx= np.where( distance[:,1:] ==0)[0]
idx
```

```
distance[idx]
```

```
index[idx]
```

```
for ix in index[idx] :
  print('x0 = ', X[ix[0]] , 'x1 = ',X[ix[1]], 'x2 = ',X[ix[2]] )
```
