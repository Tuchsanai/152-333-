```
import numpy as np
import matplotlib.pyplot as plt
import control as co

plt.rcParams['figure.figsize'] = [23,10]
plt.rcParams['font.size'] = 24
```

```
G = co.tf([2,5],[1,2,1])
G
```

```
t = np.linspace(0,9,10)
t
```

##step response

```
t = np.linspace(0,50,1000)

t1,y1 = co.step_response(G,t)

plt.plot(t1,y1)
plt.grid()
```

##Impulse response

```
t = np.linspace(0,50,1000)

t1,y1 = co.impulse_response(G,t)

plt.plot(t1,y1)
plt.grid()
```

# Force response

```
t = np.linspace(0,50,1000)
u =np.sin(np.pi*t)
```

```
t1,y1,state=co.forced_response(G,t,u)

plt.plot(t1,y1)
plt.grid()
```
